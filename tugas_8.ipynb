```python
'''
Encode target labels with value between 0 and n_classes-1
https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html
'''
from sklearn.preprocessing import LabelEncoder

for f in df.columns:
    if df[f].dtype == 'object':
        print (f)

        lbl = LabelEncoder()
        lbl.fit(list(df[f].values))

        df[f] = lbl.transform(list(df[f].values))
```

```python
X = df.drop(columns=['Cath'], axis=1) 
y = df['Cath']

X.shape, y.shape
```

```python
# Data normalization (opsional)
# Tujuan nya agar data memiliki rentang nilai yang sama
# Bandingkan performanya, lebih baik normalisasi atau tidak ??
```

```python
# Correlation Heatmap

corr = df.corr()
top_corr = corr.index

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,10))
corr_plot = sns.heatmap(df[top_corr].corr(), annot=True)
```

```python
# Seleksi Fitur / Feature Selection -> Feature Engineering
# Menggunakan Extra Tree Classifer untuk Seleksi Fitur

'''
Meta-transformer for selecting features based on importance weights.
https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SelectFromModel.html

An extra-trees classifier
https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.ExtraTreesClassifier.html
'''

from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import ExtraTreesClassifier

sel = SelectFromModel(ExtraTreesClassifier(n_estimators=100))       # model untuk seleksi fitur
sel.fit(X, y)

sel_feat = X.columns[(sel.get_support())]

print('Total fitur:', len(X.columns))
print('Jumlah fitur yang dipilih:', len(sel_feat))
print('Fitur yang dipilih:\n', sel_feat)
 
```
